1. The following code accesses the [lyrics.ovh](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search) public api, searches for the lyrics of a song, and stores it in a dictionary object. ### **Question 1 Answer:**

```python
import requests
import json

result = json.loads(requests.get('https://api.lyrics.ovh/v1/They Might Be Giants/Birdhouse in your soul').text)

# Write the JSON data to a file
with open('lyrics_data.json', 'w') as json_file:
    json.dump(result, json_file)

print("JSON data written to 'lyrics_data.json' file.")
```

2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

### **Question 2 Answer:**

```python
# Load the JSON data from the file
with open('lyrics_data.json', 'r') as json_file:
    data = json.load(json_file)

# Extract the lyrics from the JSON data
lyrics = data['lyrics']

# Print the lyrics of the song
print("Lyrics of the song:")
print(lyrics)
print()

# Perform sentiment analysis using SpaCyTextBlob
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
doc = nlp(lyrics)

# Get the polarity score
polarity_score = doc._.polarity

# Print the polarity score
print("Polarity Score:", polarity_score)

## Answer: Since the polarity score is positive, the lyrics have a positive connotation. A negative score would have a negative connotation.
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

### **Question 3 Answer:**

```python
def get_lyrics(artist, song, filename):
    # Make the API request to get the lyrics
    response = requests.get(f'https://api.lyrics.ovh/v1/{artist}/{song}')
    
    # Check if the request was successful
    if response.status_code == 200:
        # Extract the lyrics from the response
        lyrics_data = response.json()
        lyrics = lyrics_data.get('lyrics', 'Lyrics not found.')
        
        # Write the lyrics to the specified filename
        with open(filename, 'w') as file:
            file.write(lyrics)
            
        print(f"Lyrics for '{song}' by {artist} saved to '{filename}'.")
    else:
        print(f"Failed to retrieve lyrics for '{song}' by {artist}. Error {response.status_code}.")

# Test the function with four different songs and store them in different files
get_lyrics('John Legend', 'All of Me', 'all_of_me_lyrics.json')
get_lyrics('Adele', 'Someone Like You', 'someone_like_you_lyrics.json')
get_lyrics('Ed Sheeran', 'Shape of You', 'shape_of_you_lyrics.json')
get_lyrics('Queen', 'Bohemian Rhapsody', 'bohemian_rhapsody_lyrics.json')
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

### **Question 4 Answer:**

```python
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')


def calculate_sentiment_polarity(filename):
    with open(filename, 'r') as file:
        lyrics = file.read()
    
    doc = nlp(lyrics)

    sentiment_score = doc._.polarity
    return sentiment_score
# Find polarity of the files and print the polarity scores for each
files = ["all_of_me_lyrics.json", "someone_like_you_lyrics.json", "shape_of_you_lyrics.json", "bohemian_rhapsody_lyrics.json"]
for file in files:
    polarity_score = calculate_sentiment_polarity(file)
    print(f"Polarity score for '{file}: {polarity_score}")
```
